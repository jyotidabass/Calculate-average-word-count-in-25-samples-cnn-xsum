Notebook for calculating the average word count in 25 samples.

```python
!pip install datasets
from datasets import load_dataset
dataset = load_dataset('cnn_dailymail', '3.0.0')
import pandas as pd
test_df = pd.DataFrame(dataset['test'])
```

```python
# Import the necessary library
import numpy as np

# Split each article into words, count words, and store the counts in a list
word_counts = [len(article.split()) for article in test_df['article'][:25]]

# Calculate the average word count
average_word_count = np.mean(word_counts)

# Display the average word count
print(f"The average number of words in the first 25 articles of CNN dataset is {average_word_count:.2f}")
```
